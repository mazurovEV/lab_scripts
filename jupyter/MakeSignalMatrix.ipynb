```python
#Устаревший способ: смотри файл MakeCoverageInPeaks
import pandas as pd
import pickle
from pybedtools import BedTool
from multiprocessing import Pool
import pickle
import re
```

```python
def countMatrix(f):
    a = BedTool("../H3K27me3/bw/subsampleHimericBam_peaks.bed")
    b = BedTool("../H3K27me3/bw/sorted_t_pure_" + f + ".bdg")
    intersect = a.intersect(b, wao=True, sorted=True)
    return {f: countPeakSums(intersect)}
```

```python
def countPeakSums(intersect):
    signalSum = {}
    peak = ""
    for row in intersect:
        if(peak != row[3]):
            if(peak != ""): signalSum[peak] = int(partPeakSum)
            peak = row[3]
            partPeakSum = 0

        partPeakSum = partPeakSum + float(row[18])*int(row[19]) if int(row[19]) != 0 else 0
    
    return signalSum
```

```python
def atoi(text):
    return int(text) if text.isdigit() else text
```

```python
def natural_keys(text):
    return [atoi(c) for c in re.split('(\d+)', text)]
```

```python
d = pd.read_csv("../H3K27me3/H3K27me3_filtered_by_biosample.csv", sep=';', index_col=False)
files = d['File accession'].tolist()
```

```python
pool = Pool(processes=12)
matrix = pool.map(countMatrix, files)

pool.close()
pool.join()
```

```python
with open("../H3K27me3/pre_signal_matrix.pickle", 'wb') as f:
        pickle.dump(matrix, f)
```

```python
with open("../H3K27me3/pre_signal_matrix.pickle", 'rb') as f:
    data = pickle.load(f)
```

```python
data_dict = {k:v for i in data for k,v in i.items()}
```

```python
for_dataframe = {}
for k, v in data_dict.items():
    for_dataframe[k] = [data_dict[k][key] for key in sorted(data_dict[k].keys(), key=natural_keys)]
```

```python
matrix = pd.DataFrame(for_dataframe, index=sorted(data_dict['ENCFF591ZGI'].keys(), key=natural_keys))
```
