```python
def get_gene_name(gene_id):
    mg = mygene.MyGeneInfo()
    out = mg.querymany([gene_id], scopes='ensembl.gene', fields="symbol", species='human', 
                           returnall=True, as_dataframe=True)
    with_symbol = out['out']
    del with_symbol.index.name
    with_symbol['ensembl_id'] = with_symbol.index
    if 'symbol' in with_symbol.columns:
        with_symbol = with_symbol.reset_index()[['ensembl_id', 'symbol']]
        with_symbol = with_symbol[with_symbol['symbol'].notnull()]
        return with_symbol.loc[0]['symbol']
    else:
        return None
```

```python
peaks_csv = []
tissues_csv = []
genes_csv = []
for target in targets:
    print(target)
    chip = pd.read_csv("../all_marks/" + target + "/peaks_signal_matrix_norm.csv", sep="\t", index_col=0)
    ass = pd.read_csv("../all_marks/" + target + "/lncRNA_peaks_fantom_gene_association.tsv", sep="\t")
    peaks_coords = pd.read_csv("../all_marks/" + target + "/merged_peaks_first_in_biosample.bed", sep="\t", header=None)
        
    # Создаем фрейм с тканями
    tissues_csv.append(pd.DataFrame.from_records([[":".join(c.split(".")[0:2]), "Tissue"] for c in chip.columns], columns=["encode_id:ID", ":LABEL"]))
    print("tissues done")    
    # Создаем фрейм с генами и lncRNA
    # TODO: по хорошему надо бы из аннотаци гены вставлять
    new_genes_csv = ass[["gene"]]
    new_genes_csv = new_genes_csv.rename(columns={"gene": "ensembl_id:ID"})
    new_genes_csv[":LABEL"] = ["Gene"]*new_genes_csv.shape[0]
    print("genes done")    
    with h5py.File("../all_marks/" + target + "/lncRNA_Peaks_corrs/lncRNA_Peaks_Correlations_corrected_non_zero.hdf5", 'r') as f:
        lncRNAs = f['lncRNAs_names'][:]
    
    for l in list(lncRNAs):
        if l in new_genes_csv["ensembl_id:ID"]:
            new_genes_csv.at[new_genes_csv[new_genes_csv["ensembl_id:ID"] == l].index, "ensembl_id:ID"] = "Gene;lncRNA"
            lncRNAs.remove(l)
    new_genes_csv = new_genes_csv.append(pd.DataFrame.from_dict({"ensembl_id:ID": lncRNAs, ":LABEL": ["Gene;lncRNA"]*len(lncRNAs)}))        
    genes_csv.append(new_genes_csv)
    print("lncRNA genes done")
    # Файл с пиками
    peaks_coords[3] = peaks_coords[3].astype(str)
    peaks_coords = peaks_coords.rename(columns = {3: "name:string", 0: "chrom:string", 1: "start:long", 2: "end:long"})
    peaks_coords[":LABEL"] = ["Peak;Merged;" + target]*peaks_coords.shape[0]
    peaks_coords[":ID"] = target + '_' + peaks_coords["name:string"].astype(str)
    peaks_coords = peaks_coords[[":ID", "name:string", "chrom:string", "start:long", "end:long", ":LABEL"]]
    peaks_csv.append(peaks_coords)
    print("peaks done")
peaks_csv = pd.concat(peaks_csv).drop_duplicates()
tissues_csv = pd.concat(tissues_csv).drop_duplicates()
genes_csv = pd.concat(genes_csv).drop_duplicates()
```

```python
corrs_rel_csv = []
for target in targets:
    print(target)
    with h5py.File("../all_marks/" + target + "/lncRNA_Peaks_corrs/lncRNA_Peaks_Correlations_corrected_non_zero.hdf5", 'r') as f:
        shape = f['corrs_matrix'].shape

    for i in range(shape[0]):
        print(i)    
        with h5py.File("../all_marks/" + target + "/lncRNA_Peaks_corrs/lncRNA_Peaks_Correlations_corrected_non_zero.hdf5", 'r') as f:
            corrs = f['corrs_matrix'][i, :]
            lncRNA = f['lncRNAs_names'][i]

        nonzero_peak_names = ["peak_" + str(i) for i in np.nonzero(corrs)[0]]
        if len(nonzero_peak_names) > 0:
            peak_corrs = corrs[np.nonzero(corrs)[0]]
            
            corrs_rel_csv.append(pd.DataFrame.from_dict({":START_ID": [target + "_" + p for p in nonzero_peak_names], "corr:float": peak_corrs, ":END_ID": [lncRNA]*len(nonzero_peak_names), ":TYPE": ["CORR"]*len(nonzero_peak_names)}))
corrs_rel_csv = pd.concat(corrs_rel_csv)
```
