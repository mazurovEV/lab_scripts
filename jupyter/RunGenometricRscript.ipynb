```python
import rpy2.robjects as robjects
import os
import pandas as pd
from multiprocessing import Pool
```

```python
r = robjects.r
r.source("Genometric.r")
```

```python
def do_genometric(data):
    print(data)
    
    r.run_genometric(data[0], data[1], data[2], data[3], data[4])
```

```python
targets = [("H3K27ac", ""), ("H3K27me3", ""), ("H3K36me3", ""), ("H3K4me1", "_narrow"), 
           ("H3K4me2", "_narrow"), ("H3K4me3", "_narrow"), ("H3K79me2", ""), ("H3K9ac", "_narrow"), 
           ("H3K9me3", ""), ("H4K20me1", ""), ("methylation", "")]

def run_genometric(is_fantom_annotation=False):
    
    if is_fantom_annotation:
        res_folder = "fantom_genometric_margi_results"
        margi_folder = "fantom_margi"
        common_rnas_folder = "fantom_common_rnas"
    else:
        res_folder = "genometric_margi_results"
        margi_folder = "margi"
        common_rnas_folder = "common_rnas"
    
    for target in targets:
        print(target[0])

        if not os.path.exists("/home/mazurovev/all_marks/" + target[0] + "/" + res_folder + "/"):
            os.makedirs("/home/mazurovev/all_marks/" + target[0] + "/" + res_folder + "/")

        data = []
        for f in os.listdir("../all_marks/" + target[0] + "/" + margi_folder):
                data.append((f.split('.')[0], target[0], res_folder, common_rnas_folder, margi_folder))

        pool = Pool(processes=10)
        corr = pool.map(do_genometric, data)

        pool.close()
        pool.join()
```

```python
run_genometric(is_fantom_annotation=True)
```
