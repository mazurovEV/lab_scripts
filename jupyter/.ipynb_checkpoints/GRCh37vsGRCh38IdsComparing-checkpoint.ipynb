```python
from pybedtools import BedTool
#Теперь надо отсортировать этот bed и пересечь с bed с координатами пиков(на самом деле сделать closest)
targets = [("H3K27me3", "")]
for target in targets:
    fantom_anno = BedTool("../annotation/fantom_gene_annotation.bed").sort()
    peaks = BedTool("../all_marks/" + target[0] + "/merged_broad_2_peaks_first_in_biosample.bed").sort()
    closest = fantom_anno.closest(peaks, d=True).to_dataframe()
    
    anno = closest[(closest["itemRgb"]>=0)&(closest["itemRgb"]<=1000)]
    anno = anno.drop(['score', 'itemRgb'], axis=1)
    anno = anno.rename(columns={"chrom": "seqnames", "name": "feature", 
                         "start":"start_position", "end":"end_position", 
                         "strand": "start", "thickStart":"end", "thickEnd":"peak"})
    
    anno.to_csv("../all_marks/" + target[0] + "/broad_2_peaks_fantom_anno.csv", index=None, sep="\t")
```
