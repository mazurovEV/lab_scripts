```python
targets = [("H3K27ac", "_narrow"), ("H3K36me3", ""), 
           ("H3K4me1", "_narrow"), ("H3K4me2", "_narrow"), ("H3K4me3", "_narrow"), ("H3K79me2", ""), 
           ("H3K9ac", "_narrow"), ("H3K9me3", ""), ("H4K20me1", ""), ("H3K27me3", "")]#, ("methylation", "")]
```

```python
target_counts = {}
for target in targets:
    counts = {}
    i = 1
    t = pd.read_csv("../all_marks/" + target[0] + "/lncRNA_peaks_fantom_gene_association.tsv", sep="\t")
    for cluster_lncrna, cluster_genes in clus.items():
        print(i)
        hm_lncrna = t[t['lncRNA'] == cluster_lncrna]
        cluster_genes = set(cluster_genes).intersection(our_fantom_genes_common_set)
        if hm_lncrna.shape[0] > 0:
            hm_genes = set(hm_lncrna['gene']).intersection(our_fantom_genes_common_set)

            counts[cluster_lncrna] = [len(cluster_genes), len(hm_genes), len(cluster_genes.intersection(hm_genes))]
        i = i + 1
    
    target_counts[target[0]] = counts
```

```python
[(i, len([k for k, n in j.items() if n[2] > 0])) for i, j in target_counts.items()]
```

```python
with h5py.File("../all_marks/" + "H3K27ac" + "/our_fantom_fantom_genes_common_set.hdf5", 'r') as f:
    our_fantom_genes_common_set = f['set'][:]
```

```python
#TODO: в фантомовской аннотации же есть вот эти их гены CAT?
our_fantom_genes_common_set = list(set(our_fantom_genes_common_set).intersection(set(genes_for_clusters)))
```

```python
res = {}
for target, counts in target_counts.items():
    tmp = []
    p_values = []
    for rna, data in {k: v for k, v in counts.items() if v[2] != 0}.items():
        r, p = stats.fisher_exact([[data[0], data[1]], [data[2], 14747 - data[1] - data[2] + data[0]]], alternative="greater")
        if p < 0.05:
            data.append(p)
            tmp.append(rna)
            p_values.append(p)
    adjusted_pvalue = multipletests(p_values, alpha=0.05, method='fdr_bh')
    res[target] = [i for i, j in zip(tmp, adjusted_pvalue[0]) if j]
```

```python
[(target, len(rnas)) for target, rnas in res.items()]
```

```python
union_rnas = set.union(*[set(rnas) for target, rnas in res.items()])
```

```python
df_dict = {}
for r in union_rnas:
    tmp = [0]*10
    for i, target in enumerate(targets):
        if r in res[target[0]]:
            tmp[i] = 1
    df_dict[r] = tmp
```

```python
def clust_to_name(rnas):
    clusters = pd.read_csv("../fantom6_clusters/DMFB.10kb.intra_chromosomal.gene_cluster.geneID.tsv", sep="\t")
    res = []
    for r in rnas:
        res.append(clusters[clusters["ref_geneID"] == r].iloc[0]['ref_geneName'])
        
    return res
```

```python
clust_to_name(union_rnas)
```

```python
df = pd.DataFrame.from_dict(df_dict, orient="index",
                       columns=["H3K27ac", "H3K36me3", "H3K4me1", 
                                "H3K4me2", "H3K4me3", "H3K79me2", "H3K9ac", "H3K9me3", "H4K20me1", "H3K27me3"])
```
