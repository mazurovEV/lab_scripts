```python
import pandas as pd
%matplotlib inline
import seaborn as sb
from matplotlib import rcParams
import numpy as np
```

```python
chip = pd.read_csv("../all_marks/H3K27me3/peaks_signal_matrix_norm.csv", sep="\t")
```

```python
peaks = pd.read_csv("../all_marks/H3K27me3/merged_peaks_first_in_biosample.bed", sep="\t", header=None)
```

```python
data = np.log(chip.loc[peaks[peaks[0] == "chrX"][3]].T + 1)
```

```python
sb.set(color_codes=True)
rcParams['figure.figsize'] = 25,28
rcParams["patch.force_edgecolor"] = True

mask = np.zeros_like(data.values)
mask[data.values == 0] = True
with sb.axes_style("white"):
    ax = sb.heatmap(data, mask=mask)
```
